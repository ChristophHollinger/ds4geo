##Mondsee data of Magnetic Suceptibility and Density 

```
#import libraries
import pandas as pd
#plotting labrary
import matplotlib.pyplot as plt
```

```
#load data
dat_mon=pd.read_csv("https://raw.githubusercontent.com/ds4geo/ds4geo_ws2020/master/student_submitted_data/ChristophDaxer/Mon2016_MagSus_Density.csv?token=ARIFOUGWJJAZH47B45UDTX27ROIJG")
```

```
#view table
dat_mon
```

```
# rename columns 
dat_mon.columns=["depth_m","suscept","depth_d","density"]
print(dat_mon.columns)
```

```
# Plot susceptibility 
plt.plot(dat_mon["suscept"],dat_mon["depth_m"])
plt.xlabel("Magnetic Susceptibility (SI x 10^-5)")
plt.ylabel("Depth (cm)")
```

```
# Plot density 
plt.plot(dat_mon["density"],dat_mon["depth_d"])
plt.xlabel("Density (g/cm^3)")
plt.ylabel("Depth (cm)")
```

```
# Plot dataset Mondsee

fig,ax=plt.subplots()
ax2=ax.twinx()
ax2=ax.twiny()
ax.plot(dat_mon["suscept"],dat_mon["depth_m"],color="orange",label='depth_m')
ax2.plot(dat_mon["density"],dat_mon["depth_d"],color="blue",label='depth_d')
ax.set_xlabel("Magnetic Susceptibility (SI x 10^-5)")
ax2.set_xlabel("Density (g/cm^3)")
ax.set_ylabel("Depth for Susceptibility (cm)")
ax2.set_ylabel("Depth for Density (cm)")

```
